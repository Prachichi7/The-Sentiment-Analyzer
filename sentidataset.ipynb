```python
grouping= df.groupby(['Age of User','sentiment']).size().unstack(fill_value=0)
print(grouping)
```

```python
#corr= df.corr(grouping)
import matplotlib.pyplot as plt

grouping.plot(kind='bar', stacked=True, figsize=(10,6), color=['black', 'skyblue', 'grey'])
plt.title("Sentiment Distribution by Age")
plt.xlabel("Age of User")
plt.ylabel("Count")
plt.show()
```
